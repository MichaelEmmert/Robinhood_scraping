```python
#data cleaning
Robinhood_Stocks = pd.read_csv('C:/Users/Micha/Desktop/JupyterNotebooks/Robinhood/Robinhood/Robinhood_stocks_6242020.csv')

#BUY HOLD AND SELL
Robinhood_Stocks.Buy_Hold_Sell = Robinhood_Stocks.Buy_Hold_Sell.str.replace('%','')
#BUY
Robinhood_Stocks['Buy'] = Robinhood_Stocks.Buy_Hold_Sell\
[Robinhood_Stocks.Buy_Hold_Sell.notnull()].apply(lambda x: int(x.split(',')[0]))

#HOLD
Robinhood_Stocks['Hold'] = Robinhood_Stocks.Buy_Hold_Sell\
[Robinhood_Stocks.Buy_Hold_Sell.notnull()].apply(lambda x: int(x.split(',')[1]))

#SELL
Robinhood_Stocks['Sell'] = Robinhood_Stocks.Buy_Hold_Sell\
[Robinhood_Stocks.Buy_Hold_Sell.notnull()].apply(lambda x: int(x.split(',')[2]))

#Market Price
Robinhood_Stocks.Market_Price = Robinhood_Stocks.Market_Price.str.replace(',','').str.replace('$','').astype(float)

# collections
Robinhood_Stocks.Collections = Robinhood_Stocks.Collections\
[Robinhood_Stocks.Collections.notnull()].apply(lambda x: x.split(','))

#ratings
Robinhood_Stocks.Ratings = Robinhood_Stocks.Ratings[Robinhood_Stocks.Ratings.notnull()]\
.apply(lambda x: x.split(',')[1]).apply(lambda x: int(x.split(' ')[1]))

#Employees
Robinhood_Stocks.Employees[Robinhood_Stocks.Employees == '—'] = np.NaN
Robinhood_Stocks.Employees = Robinhood_Stocks.Employees[Robinhood_Stocks.Employees.notnull()]\
.str.replace(',','').astype(float)

#Market_cap

#PE_Ratio
#Dividend Yield
# Robinhood_Stocks.Dividend_Yield[Robinhood_Stocks.Dividend_Yield == '—'] = np.NaN
# Robinhood_Stocks.Dividend_Yield = Robinhood_Stocks.Dividend_Yield.astype(float)
```

```python
Robinhood_Stocks = Robinhood_Stocks[['Company_Name','Symbol','Market_Price','Employees',
                                     'Headquarters','Dividend_Yield','Market_Cap','PE_Ratio',
                                     'Ratings','Buy','Hold','Sell','Collections','Average_Volume']]
Robinhood_Stocks.head()
```
